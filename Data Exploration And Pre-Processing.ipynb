```python
q1 = 321950
q3 = 645000
```

```python
iqr = q3 - q1
print("The Inter-Quartile Range Is: ", iqr)
```

```python
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr


print("The Lower Limit Is ", lower_limit)
print("The Upper Limit Is ", upper_limit)
```

```python
def limit_imputer(value):
  if value > upper_limit:
    return upper_limit
  if value < lower_limit:
    return lower_limit
  else:
    return value
```

```python
data['Sale Price'] = data['Sale Price'].apply(limit_imputer)

#OUTLIERS HAVE BEEN REMOVED

data['Sale Price'].describe()
```

# **`CATEGORICAL VARIABLE DATA EXPLORATION`**

**`ANOVA`**

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
data = data.rename(columns= {'Sale Price' : 'Sale_Price'})
data = data.rename(columns= {'Condition of the House' : 'Condition_of_the_House'})
data = data.rename(columns= {'Waterfront View' : 'Waterfront_View'})
data = data.rename(columns= {'Ever Renovated' : 'Ever_Renovated'})
```

```python
data.head()
```

```python
#ANOVA
#Checks whether the variable has an impact on the target variable


mod = ols('Sale_Price ~ Condition_of_the_House', data = data).fit()
sm.stats.anova_lm(mod, typ = 2)
```

```python
#Create Dummy Variables

data = pd.get_dummies(data, columns= ['Condition_of_the_House'], drop_first= True)
```

```python
data = pd.get_dummies(data, columns= ['Waterfront_View', 'Ever_Renovated'], drop_first= True)
```

# **`BINNING`**

```python
zip = data.groupby('Zipcode').agg({'Sale_Price': 'mean'}).sort_values('Sale_Price', ascending = True)
zip.head()
```

```python
zip['Zip_Groups'] = pd.cut(zip['Sale_Price'], bins = 10,
                          labels = ['Zip_0',
                                    'Zip_1',
                                    'Zip_2',
                                    'Zip_3',
                                    'Zip_4',
                                    'Zip_5',
                                    'Zip_6',
                                    'Zip_7',
                                    'Zip_8',
                                    'Zip_9'], include_lowest = True)
```

```python
zip = zip.drop(columns = 'Sale_Price')
```

```python
data = pd.merge(data, zip, left_on= 'Zipcode', how = 'left', right_index= True)
```

```python
data = data.drop(columns = 'Zipcode')
```

```python
data = pd.get_dummies(data, columns= ['Zip_Groups'], drop_first= True)
```
