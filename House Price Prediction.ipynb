```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data = pd.read_csv("/content/Transformed Housing Data.csv")
data.head()
```

# **`SCALING`**




```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

Y = data['Sale_Price']
X = sc.fit_transform(data.drop( columns = ['Sale_Price']))
```

```python
X = pd.DataFrame( data = X, columns = data.drop(columns = ['Sale_Price']).columns)
X.head()
```

# **`CHECKING AND REMOVING MULTI-COLLINEARITY`**

```python
X.corr()
```

```python
k = X.corr()
z = [[str(i), str(j)] for i in k.columns for j in k.columns if(k.loc[i,j]>abs(0.5))&(i!=j)]
z
```

# **`CALCULATING VIF`**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_data = X

VIF = pd.Series([vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
VIF[VIF== VIF.max()].index[0]
```

```python
def MC_remover(data):
  VIF = pd.Series([vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
  if VIF.max() > 5:
    print(VIF[VIF == VIF.max()].index[0], 'has been removed')
    data = data.drop(columns = VIF[VIF== VIF.max()].index[0])
    return data
  else:
    print("NO MULTICOLLINEARITY")
    return data
```

```python
for i in range(7):
  vif_data = MC_remover(vif_data)

vif_data.head()
```

```python
VIF = pd.Series([vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

# **`TRAIN TEST SETS`**

```python
X = vif_data
Y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2)
```

# **`MODEL`**

```python
from traitlets.config.application import T
# LINEAR REGRESSION MODEL

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr.coef_
```

```python
predictions = lr. predict(X_test)
predictions
```

```python
# Accuracy

lr.score(X_test, Y_test)
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': predictions})
df1 = df.head(10)
df1
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
print('VarScore:',metrics.explained_variance_score(Y_test,predictions))
```
