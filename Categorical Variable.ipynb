```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("/content/Transformed_Housing_Data.csv")
data.head()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
data = data.rename(columns= {'Sale Price' : 'Sale_Price'})
data = data.rename(columns= {'Condition of the House' : 'Condition_of_the_House'})
data = data.rename(columns= {'Waterfront View' : 'Waterfront_View'})
data = data.rename(columns= {'Ever Renovated' : 'Ever_Renovated'})
```

```python
data.head()
```

```python
#ANOVA
#Checks whether the variable has an impact on the target variable


mod = ols('Sale_Price ~ Condition_of_the_House', data = data).fit()
sm.stats.anova_lm(mod, typ = 2)
```

```python
#Create Dummy Variables

data = pd.get_dummies(data, columns= ['Condition_of_the_House'], drop_first= True)
```

```python
data = pd.get_dummies(data, columns= ['Waterfront_View', 'Ever_Renovated'], drop_first= True)
```

```python
data.head()
```

# **`BINNING`**

```python
zip = data.groupby('Zipcode').agg({'Sale_Price': 'mean'}).sort_values('Sale_Price', ascending = True)
zip.head()
```

```python
zip['Zip_Group'] = pd.cut(zip['Sale_Price'], bins = 10,
                          labels = ['Zip_0',
                                    'Zip_1',
                                    'Zip_2',
                                    'Zip_3',
                                    'Zip_4',
                                    'Zip_5',
                                    'Zip_6',
                                    'Zip_7',
                                    'Zip_8',
                                    'Zip_9'], include_lowest = True)
```

```python
zip = zip.drop(columns = 'Sale_Price')
```

```python
data = pd.merge(data, zip, left_on= 'Zipcode', how = 'left', right_index= True)
```

```python
data = data.drop(columns = 'Zipcode')
```

```python
data = pd.get_dummies(data, columns= ['Zip_Group'], drop_first= True)
```
